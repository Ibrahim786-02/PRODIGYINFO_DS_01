```python
import pandas as pd
import matplotlib.pyplot as plt
afghanistan_data = data[data['Country Name'] == 'Afghanistan'].iloc[0, 4:-2]
afghanistan_data = pd.to_numeric(afghanistan_data, errors='coerce').dropna()
plt.figure(figsize=(16, 8))
plt.bar(afghanistan_data.index, afghanistan_data.values, color='red', edgecolor='black')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.title('Population of Afghanistan (1960-2022)', fontsize=16)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
plt.text(afghanistan_data.index[0], afghanistan_data.values[0], f'{afghanistan_data.values[0]:,.0f}', 
         ha='center', va='bottom', fontsize=10, color='blue')
plt.text(afghanistan_data.index[-1], afghanistan_data.values[-1], f'{afghanistan_data.values[-1]:,.0f}', 
         ha='center', va='bottom', fontsize=10, color='blue')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
countries = ['Afghanistan', 'India', 'United States', 'China', 'Brazil']
population_2020 = data[data['Country Name'].isin(countries)][['Country Name', '2020']]
population_2020['2020'] = pd.to_numeric(population_2020['2020'], errors='coerce')
plt.figure(figsize=(14, 7))
bars = plt.bar(population_2020['Country Name'], population_2020['2020'], color=['blue', 'green', 'red', 'purple', 'orange'], edgecolor='black')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Population in 2020', fontsize=14)
plt.title('Population Comparison of Selected Countries in 2020', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:,.0f}', ha='center', va='bottom', fontsize=10, color='black')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
countries = ['Afghanistan', 'India', 'United States', 'China', 'Brazil']
population_data = data[data['Country Name'].isin(countries)].set_index('Country Name')
years = population_data.columns[4:-2]
plt.figure(figsize=(14, 7))
for country in countries:
    country_data = pd.to_numeric(population_data.loc[country, years], errors='coerce').dropna()
    plt.plot(country_data.index, country_data.values, marker='o', label=country)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.title('Population Growth Over Time for Selected Countries (1960-2022)', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
def handle_missing_values(df):
    # Drop columns with all NaN values
    df = df.dropna(axis=1, how='all')
    # Convert to numeric, coercing errors to NaN
    df = df.apply(pd.to_numeric, errors='coerce')
    # Replace remaining NaNs with the mean of the column
    df = df.fillna(df.mean())
    return df
region = 'Africa Eastern and Southern'
region_data = data[data['Country Name'] == region]
population_columns = [col for col in region_data.columns if col.isdigit() and int(col) >= 1960]
region_data = region_data[population_columns]
region_data = handle_missing_values(region_data.copy())  # Avoid modifying original data
average_population = region_data.mean(axis=1).mean()
plt.figure(figsize=(10, 6))
plt.bar([region], [average_population], color='teal', edgecolor='black')
plt.xlabel('Region', fontsize=14)
plt.ylabel('Average Population (1960-2022)', fontsize=14)
plt.title('Average Population of Africa Eastern and Southern (1960-2022)', fontsize=16)
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
countries = ['Afghanistan', 'India', 'United States', 'China', 'Brazil']
population_data = data[data['Country Name'].isin(countries)].set_index('Country Name')
years = population_data.columns[4:-2]
colors = ['skyblue', 'green', 'red', 'purple', 'orange']
fig, axes = plt.subplots(len(countries), 1, figsize=(12, 10), sharex=True)
for i, country in enumerate(countries):
    country_data = pd.to_numeric(population_data.loc[country, years], errors='coerce').dropna()
    axes[i].hist(country_data.values, bins=20, alpha=0.7, color=colors[i], edgecolor='black')
    axes[i].set_title(f'Population Distribution for {country} (1960-2022)', fontsize=12)
    axes[i].set_xlabel('Population', fontsize=10)
    axes[i].set_ylabel('Frequency', fontsize=10)
    axes[i].grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```
